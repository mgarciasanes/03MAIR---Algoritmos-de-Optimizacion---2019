```
#quick sort


import random
A=random.sample(range(1, 1000), 20)


#Calculo de tiempo de ejecución
from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):   
    def wrapper(*args, **kwargs):        
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    return wrapper
  

def quick_sort(lista_a_ordenar):

  if len(lista_a_ordenar)==1:
    return lista_a_ordenar

  if len(lista_a_ordenar)==2:
    return [min(lista_a_ordenar),max(lista_a_ordenar)]
  
  I=[]
  D=[]
  
  pivote = (lista_a_ordenar[0]+lista_a_ordenar[1]+lista_a_ordenar[2])/3
  
  for i in lista_a_ordenar:
    if i<pivote:
      I.append(i)
    else:
      D.append(i)
      
  return quick_sort(I) + quick_sort(D)                

print(A)

@calcular_tiempo
def qs(lista_a_ordenar):
    print(quick_sort(lista_a_ordenar))
    
    
qs(A)

  
```

```
@calcular_tiempo
def cambio_monedas(CANTIDAD,SISTEMA=[25,10,5,1]):
  SOLUCION=[0 for i in range(len(SISTEMA))]
  VALOR_ACUMULADO=0
  for i in range(len(SISTEMA)):
      monedas=int((CANTIDAD - VALOR_ACUMULADO)/SISTEMA[i])
      
      SOLUCION[i] = monedas
      
      VALOR_ACUMULADO += monedas*SISTEMA[i]
      
      if CANTIDAD == VALOR_ACUMULADO:
        return SOLUCION
      
cambio_monedas(99)
```

```
N=8
solucion=[0 for i in range(N)]
etapa=0


def es_prometedora(solucion, etapa):
  for i in range(etapa+1):
    if solucion.count(solucion[i]) >1 :
      return False
    for j in range (i+1, etapa+1):
      if abs(i-j) == abs(solucion[i]-solucion[j]):
        return False
  return True
      
      
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if solucion[i]==x+1:
        print(" X ", end="")
      else:
        print(" - ", end="")
  
  
def reinas(N,solucion, etapa):
  for i in range(1,N+1):
    solucion[etapa]=i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print("\n\nLa solucion es:")
        print(solucion)
        escribe(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None
    solucion[etapa]=0
    
reinas(N,solucion,etapa)
    
    
```
