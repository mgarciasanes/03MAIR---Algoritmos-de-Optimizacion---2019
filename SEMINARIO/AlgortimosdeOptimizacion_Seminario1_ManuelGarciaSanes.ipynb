## Posibles soluciones sin restricciones:


El problema consiste en agrupar el ganado de forma que el peso en cada grupo sea lo mas homogeneo posible. Adicionalmente, tambien se nos presenta un problema simple de clasificacion por genero entre vacas hembra y macho. 


Necesitamos entonces 9 grupos, porcada clasificacion de genero.

Digamos que los grupos tienen la siguiente formar:
<br>

grupo 1  -->  {x<sub>1</sub>,x<sub>2</sub>,x<sub>3</sub>,.....x<sub>n1</sub>} <br>
grupo 2  -->  {x<sub>n1+1</sub>,.....x<sub>n2</sub>} <br>

El numero de grupos es conocido, L=9.  Para efectos de calculo de posibilidades asumimos que todos los grupos tendran el mismo numero de vacas, siendo este:
<br>
<br>
**N<sub>m</sub>/L = n = n<sub>1</sub> = n<sub>2</sub>/2 = n<sub>3</sub>/3...**

Segun la formula de  combinatoria, con N<sub>m</sub> elementos, podemos obtener un total de combinaciones de n sub grupos:
<br>
<br>
**ùê∂(N<sub>m</sub>,n) = N<sub>m</sub>!/(n!(N<sub>m</sub>‚àín)!)**
<br>

Pero aqui no acaba el problema, la formula anterior solo nos dice cuantas grupos de n elementos son posibles formar a partir del total N<sub>m</sub>. Sin embargo, nosotros necesitamos 9 grupos.  Esto genera
<br>
<br>

**Total_Parcial = ùê∂(N<sub>m</sub>,n) x ùê∂(N<sub>m</sub>-n,n) x ùê∂(N<sub>m</sub>-2n,n) x  ...  x ùê∂(N<sub>m</sub>-8n,n)**

<br>
Al "Total_Parcial", aun tenemos que quitarle las posibilidades duplicadas, pero el orden es de todas formas **factorial**.

## Posibles soluciones con restricciones:

- Aunque el numero de grupos esta definido, la cantidad de vacas en cada grupo sera flexible. El valor de **delta** se utilizara como variable  cuya variacion adiciona diferentes posibilidades de tamanos en cada grupo. (Se explica mas sobre esta variable en las siguientes secciones)

## Modelo de Espacio de Soluciones

Despues de un analisis del problema, resulta evidente que al ordenar los datos en funcion de la variable de interes, en este caso el **peso**,  automaticamente estamos juntando  los valores  cercanos entre si. El problema a solucionar entonces es el encontrar el tamano de cada grupo de forma que la **varianza** sea minima. Mientras menor sea la varianza, mas homogeneo es el grupo. 

Siendo que los grupos son de la forma:
<br>
<br>
grupo 1  -->  {x<sub>1</sub>,x<sub>2</sub>,x<sub>3</sub>,.....x<sub>n1</sub>} <br>
grupo 2  -->  {x<sub>n1+1</sub>,.....x<sub>n2</sub>} <br>
grupo 3  -->  {x<sub>n2+1</sub>,.....x<sub>n3</sub>} <br>
...<br>
grupo 9 --> {x<sub>n8+1</sub>,.....x<sub>n9</sub>} <br>
<br>
donde x<sub>1</sub> < x<sub>2</sub> <  x<sub>3</sub> <  ... < x<sub>n1</sub> < ... < x<sub>n9</sub> <br>
<br>
la solucion a encontrar, tendra la forma:

**SOLUCION = {n1, n2, n3, n4, n5, n6, n7, n8, n9}**

Donde n1,n2,...n9 son indices de posicion de los datos x<sub>n1</sub>, x<sub>n2</sub>...x<sub>n9</sub> respectivamente. 
<br>


**Nota:** Inicialmente, se intento obtener como solucion los grupos directamente. Pero dada la complejidad del problema, quedo evidente que la solucion es inviable.

## Funcion Objetivo

Lo que intentamos lograr, es homogeneidad en cada grupo. Para tal efecto, debemos de **minimizar** la sumatoria de las varianzas de cada grupo

 **VAR<sub>Total</sub> = SUM<sub>j=1-->L</sub>  [   1/t<sub>j</sub> x SUM<sub>i=0-->n<sub>j</sub></sub>   [ ( x<sub>i</sub> - X<sub>j</sub>)<sup>2</sup> ]   ] **

donde: <br>
n<sub>j</sub> es el ultimo dato del grupo j <br>
L es el numero de grupos
X<sub>j</sub> es el promedio de los valores del grupo j<br>
t<sub>j</sub> es el numero de datos en el grupo j <br>

## Maximizacion o minimizacion

Es un caso de **minimizacion** de la varianza

## Algoritmo de busqueda: Fuerza Bruta

La celda de codigo a continuacion contiene las siguientes funciones:

- **podar**:  Esta funcion poda todos los nodos cuyas cotas inferior sean mayor a la mejor (menor) cota superior disponible (variable de cota)

- **mejor_nodo**: Esta funcion encuentra el nodo con mejor (menor) coste (valor de BE) y lo devuelve para ser explorado

- **siguiente_nodo**: Esta fucion retorna el siguiente nodo a explorar. Se usa para "FUERZA BRUTA"

- **es_solucion**: Esta funcion verifica si es que el nodo en el ultimo nivel del arbol es una solucion optima

- **print_debug**: Esta funcion imprime mensajes en pantalla

- **estimar_cotas**: Esta funcion calcula los siguientes valores:
 - ci: Cota inferior<br>
  - cs: Cota superior<br>
 -  be: Costo/Beneficio <br>
 -  parcial: Calculo de la funcion objetivo hasta este punto<br>
  
  
- **vaquitas_a_mi**: Esta fucion es una implementacion del algoritmo de busqueda "RAMIFICACION Y PODA". **Pero es configurable de modo que puede realizar la busqueda sin realizar la poda, lo que puede considerarse "FUERZA BRUTA"**. En otras palabras, es un arbol de busqueda que calcula y analiza todos los nodos. 
<br>La funcion incluye comentarios con instrucciones para su uso. 

```
import random
import numpy as np
import pprint
import sys
import pandas as pd
sys.setrecursionlimit(100000)
import math


def podar(nodos_todos,debug_mode=False):
  # Esta funcion poda todos los nodos cuyas cotas inferior es mayor a la mejor cota superior disponible (variable de cota)
  variable_poda=100000000000
  nodo_variable_poda={}
  cantidad=len(nodos_todos)
  for n in nodos_todos:
    if n['cs']<variable_poda:
      variable_poda=n['cs']
      nodo_variable_poda=n
  print_debug("<<<<<<<<<<<<<< Podando Nodos -->  Variable de Poda: " + str(variable_poda) + " de nodo con id " + str(nodo_variable_poda['id'])+ " y sec:  " +str(nodo_variable_poda['secuencia']),debug_mode=debug_mode)
  for n in nodos_todos:
    if n['ci']>variable_poda:
      print_debug("<<<<<<<<<<<<<< Podando Nodo con id: " + str(n['id']) + " con CS " + str(n['cs']) + " con CI " + str(n['ci']) + " con sec  " + str(n['secuencia'])  + " en nivel " + str(n['nivel']),debug_mode=debug_mode)
      nodos_todos.remove(n)
  return nodos_todos


def mejor_nodo(nodos_todos,L):
  # Esta funcion encuentra el nodo con mejor (menor) coste (valor de BE) y lo devuelve para ser explorado
  if len(nodos_todos)==1:
    return nodos_todos[0],[]
  if len(nodos_todos)==0:
    return [],[]
  mejor_nodo_en_registro=[]
  mejor_be=10000000000000000
  nivel_contador=0
  
  for n in nodos_todos:
    if n['nivel']==L:
      nivel_contador=nivel_contador+1
      continue
    if n['be']<mejor_be:
      mejor_be=n['be']
      mejor_nodo_en_registro=n
  if mejor_nodo_en_registro!=[]:
    nodos_todos.remove(mejor_nodo_en_registro)
  else:
    if nivel_contador==len(nodos_todos):
      return mejor_nodo(nodos_todos,-1)    
  return mejor_nodo_en_registro,nodos_todos



def siguiente_nodo(nodos_todos):
  # Esta fucion retorna el siguiente nodo a explorar. Se usa para "FUERZA BRUTA"
  if len(nodos_todos)==1:
    return nodos_todos[0],[]
  if len(nodos_todos)==0:
    return [],[]
  mejor_nodo=nodos_todos[-1]
  nodos_todos.remove(mejor_nodo)
  return mejor_nodo,nodos_todos


def es_solucion(nodos_nivel,nodos):
  # Esta funcion verifica si es que el nodo en el ultimo nivel del arbol es una solucion optima
  potenciales=[]
  for nodo_nivel in nodos_nivel:
    potencial=True
    for nodo in nodos:
      if nodo_nivel['cs']>nodo['ci']:
        potencial=False
    if potencial:
      potenciales.append(nodo_nivel)
  return mejor_nodo(potenciales,-1)
  

def print_debug(msg,pretty=False,debug_mode=False):
  #Esta funcion imprime mensajes en pantalla
  if not debug_mode:
    return 
  if pretty:
    pprint.pprint(msg)
  else:
    if isinstance(msg,list):
      for i in msg:
        print(i)
    else:
      print(msg)
    
def estimar_cotas(data,data_indice,L,padre):
  #Esta funcion calcula los siguientes valores:
  # ci: Cota inferior
  # cs: Cota superior
  # be: Costo/Beneficio 
  # parcial: Calculo de la funcion objetivo hasta este punto

  overlap=1
  niveles_pendientes=L-(padre['nivel']+1)
  datos_pendientes=len(data[data_indice:])
  pasos=datos_pendientes//niveles_pendientes
  caso_maximo=0
  caso_minimo=0
  for nn in range(niveles_pendientes):
    if nn == niveles_pendientes-1:
      caso_maximo=caso_maximo+np.var(data[data_indice+pasos*nn-overlap:])
      caso_minimo=caso_minimo+np.var(data[data_indice+pasos*nn+overlap:])
    elif nn==0:
      caso_maximo=caso_maximo+np.var(data[data_indice+pasos*nn:data_indice+pasos*(nn+1)])
      caso_minimo=caso_minimo+np.var(data[data_indice+pasos*nn:data_indice+pasos*(nn+1)-overlap])      
    else:
      caso_maximo=caso_maximo+np.var(data[data_indice+pasos*nn-overlap:data_indice+pasos*(nn+1)])
      caso_minimo=caso_minimo+np.var(data[data_indice+pasos*nn+overlap:data_indice+pasos*(nn+1)-overlap])

  if padre['value']==-1:
    cs=caso_maximo
    ci=caso_minimo
    parcial=0
    be=(ci+cs)/2
  else:
    parcial=np.var(data[padre['value']:data_indice])+padre['parcial']
    ci=parcial+caso_minimo
    cs=parcial+caso_maximo
    be=(ci+cs)/2
    parcial=parcial
  return cs,ci,be,parcial


def vaquitas_a_mi(data,L,delta=1,poda=True,debug_mode=False,nodo_cuenta=0,nodos=[],padre={},fuerza_bruta=[]):
  #  Esta funcion es la implementacion de los siguientes algoritmos:
  #  poda: True  -->  ARBOL DE BUSQUEDA: RAMIFICACION Y PODA
  #  poda: False -->  ARBOL DE BUSQUEDA: FUERZA BRUTA
  #     L: Es la cantidad de grupos homogeneos que tenemos que conseguir
  # delta: Variable que define el rango de diferentes tamanos cada grupo puede tomar
  #        si n es igual a N//L  donde N es la cantidad de datos y L el numero de grupos 
  #        entonces el tamano de los  grupos estaran definidos como:
  #        grupo 1  [x0,x1,...,{n-delta:n+delta}]
  #        grupo 2  [{n-delta:n+delta}+1,...{n*2-delta:n*2+delta}]
  #        ...
  #        grupo 9  [{n*8-delta:n*8+delta}+1,...xN] 
  #
  #        Por Ejemplo:
  #
  #        Si L=9, N=95, delta=1, n=95//9=10
  #
  #        Tenemos 3 posibilidades para grupo 1  
  #        Esto se traslada en 3 nodos en el primer nivel del arbol  (2*delta+1 Nodos) 
  #        nodo1[value=9]  --> grupo1=[x0,x1,...,x9]   (9=10*1-1)
  #        nodo2[value=10] --> grupo1=[x0,x1,...,x10]  (10=10*1)
  #        nodo3[value=11] --> grupo1=[x0,x1,...,x11]  (11=10*1+1)
  #       
  #        siguiendo en el nivel 2 del arbol, obtenemos 3 (3=2*delta+1) nodos por cada nodo en nivel 1
  #
  #        del nodo1,  nodo4[value=19] -->  grupo2=[x10,x11,...,x19]  (19=10*2-1)
  #        del nodo1,  nodo5[value=20] -->  grupo2=[x10,x11,...,x20]  (20=10*2)
  #        del nodo1,  nodo6[value=21] -->  grupo2=[x10,x11,...,x21]  (21=10*2+1)
  #        del nodo2,  nodo7[value=19] -->  grupo2=[x11,x12,...,x19]  (19=10*2-1)
  #        del nodo2,  nodo8[value=20] -->  grupo2=[x11,x12,...,x20]  (20=10*2)
  #        del nodo2,  nodo9[value=21] -->  grupo2=[x11,x12,...,x21]  (21=10*2+1)
  #        del nodo3,  nodo10[value=19] --> grupo2=[x12,x13,...,x19]  (19=10*2-1)
  #        del nodo3,  nodo11[value=20] --> grupo2=[x12,x13,...,x20]  (20=10*2)
  #        del nodo3,  nodo12[value=21] --> grupo2=[x12,x13,...,x21]  (21=10*2+1)
  #
  #        y asi sucesivamente...
  #
  # debug_mode: Activa o desactive el modo "debug" el cual genera mensajes en pantalla para cada parte del algoritmo
  # nodo_cuenta,nodos, padre,fuerza_bruta son argumentos utilizados para recursividad. 
  #
  # Esta funcion retorna el nodo que contiene la solucion optima y el numero de nodos que se han calculado en total.
  
  #Generamos el nodo 0, nodo raiz.
  if padre=={}:
    padre={'id':0,'nivel':0,'secuencia':[],'be':0,'ci':0,'cs':0,'parcial':0,'value':-1}
    padre['cs'],padre['ci'],padre['be'],padre['parcial']=estimar_cotas(data,0,L,padre)                 
    padre['value']=0
    
  print_debug("################### Nivel " +str(padre['nivel']+1) + ". Explorando Nodo " + str(padre) ,debug_mode=debug_mode)
 
    
  #Inicializamos valores
  # N: Numero de datos total
  # n: Numero de datos inicial en cada grupo
  # nodo_nivel: Lista que contiene los nuevos nodos descubiertos en esta ronda
  N=len(data)
  n=(N//L)
  nodo_nivel=[]

  #Para todos los Niveles del Arbol, menos el ultimo. Abrimos NODOS!
  if padre['nivel']<L-1:
    #Un nodo para cada posible valor del limite superior.  n --> [n*nivel-delta:n*nivel+delta]
    ni=n*(padre['nivel']+1)-delta
    nf=n*(padre['nivel']+1)+delta+1
    for i in range(ni,nf):
      #...evitar valores negativos en los indices
      if (i-padre['value']-1)<=1:
        continue
      #Crear NODO
      node={}
      #ID
      nodo_cuenta=nodo_cuenta+1
      node['id']=nodo_cuenta
      #Calcular Cota Inferior, Cota Superior,Coste (BE) y el calculo de la varianza hasta este punto en el arbol (parcial)
      cs,ci,be,parcial=estimar_cotas(data,i,L,padre)
      node['ci']=ci
      node['cs']=cs
      node['be']=be
      node['parcial']=parcial
      #valor del indice
      node['value']=i
      #Nodo Padre
      node['padre']=padre['id']
      #secuencia de indices hasta este punto en el arbol
      node['secuencia']=padre['secuencia']+[i]
      #Nivel en el arbol
      node['nivel']=padre['nivel']+1
      #almacenar Nodo en lista temporal
      nodo_nivel.append(node)

  #Ultimo Nivel del Arbol    
  if padre['nivel']==L-1:
    if (N-padre['value']-1)>1:
      node={}
      nodo_cuenta=nodo_cuenta+1
      node['id']=nodo_cuenta
      node['value']=N-1
      #En el ultimo nivel del arbol, ci=cs=be=parcial
      node['ci']=np.var(data[padre['value']+1:])+padre['parcial']
      node['cs']=node['ci']
      node['be']=node['ci']
      node['parcial']=node['ci']
      node['padre']=padre['id']
      node['nivel']=padre['nivel']+1
      node['secuencia']=padre['secuencia']+[N-1]
      
      if poda:
        #RAMIFICACION Y PODA
        nodo_nivel.append(node)
        #Verificar si la solucion en este nodo es solucion optima
        nodo_mejor,resto_nodos=es_solucion(nodo_nivel[:],nodos[:])
        if nodo_mejor!=[]:
          print_debug("********************  Es ultimo nivel. Nodo es optimo:",debug_mode=debug_mode)
          print_debug(nodo_mejor,debug_mode=debug_mode)
          return nodo_mejor,nodo_cuenta
        else:
          print_debug("********************  Es ultimo nivel. Este nodo no es solucion...",debug_mode=debug_mode)
          print_debug(nodo_nivel[:],debug_mode=debug_mode)
          pass
      else:
        #FUERZA BRUTA
        print_debug("********************  Es ultimo nivel en modo Fuerza Bruta.",debug_mode=debug_mode)
        fuerza_bruta.append(node)

        
  if padre['nivel']==L:
    print_debug("********************  Regresando a una solucion anterior. Este nodo no es solucion...",debug_mode=debug_mode)
    print_debug(padre,debug_mode=debug_mode)
    return padre,nodo_cuenta
       
  #Mostrar progreso (si es que debug_mode es True)
  print_debug("Revisando nodos hasta este punto",debug_mode=debug_mode)      
  print_debug("******************* Nodos activos:",debug_mode=debug_mode)
  print_debug(nodos,debug_mode=debug_mode)
  print_debug(">>>>>>>>>>>>>>>>>>> Abriendo Nodos -- Nuevos nodos en este nivel:",debug_mode=debug_mode)
  print_debug(nodo_nivel,debug_mode=debug_mode)
    
    
  if not poda:
    #FUERZA BRUTA
    #Mostrar los nodos del ultimo nivel del arbol hasta este punto (si es que debug_mode es True)
    print_debug("~~~~~~~~~~~~~~~~~~~~ Nodos Fuerza Bruta:",debug_mode=debug_mode)
    print_debug(fuerza_bruta,debug_mode=debug_mode)
  
  
  if poda:
    #RAMIFICACION Y PODA
    #Podar...
    nodos=podar(nodos[:]+nodo_nivel[:],debug_mode=debug_mode)
    #Elegir mejor nodo (con menor BE), para explorar...
    nodo_mejor,resto_nodos=mejor_nodo(nodos[:],L)
  else:
    #FUERZA BRUTA
    #obtener siguiente Nodo de la pila
    nodo_mejor,resto_nodos=siguiente_nodo(nodos[:]+nodo_nivel[:])
    
  #Mostrar progreso (si es que debug_mode es True)
  print_debug("----------------------Siguiente nodo a explorar:",debug_mode=debug_mode)
  print_debug(nodo_mejor,debug_mode=debug_mode)
  print_debug("",debug_mode=debug_mode)
  print_debug("",debug_mode=debug_mode)


  if nodo_mejor!=[]:
      #Si es que tenenmos Nodo para explorar, exploremos...
      return vaquitas_a_mi(data,L,delta=delta,poda=poda,debug_mode=debug_mode,nodo_cuenta=nodo_cuenta,nodos=resto_nodos,padre=nodo_mejor,fuerza_bruta=fuerza_bruta)
  else:
    #No hay mas nodos para explorar...
    if poda:
      #RAMIFICACION Y PODA
      #Teoricamente, no tendriamos que llegar a este punto
      print_debug("Nos quedamos sin nodos, algo paso... :",debug_mode=debug_mode)
      print_debug("Se acabo!",debug_mode=debug_mode)
    else:
      #FUERZA BRUTA
      #Se han analizado todos los nodos
      #Elegir el que tiene menor varianza y retornanrlo como respuesta
      print_debug("Fuerza bruta: Llegamos al final. Calculando solucion:",debug_mode=debug_mode)
      menor_var=100000000000
      nodo_mejor={}
      for i in range(len(fuerza_bruta)):
        if fuerza_bruta[i]['parcial']<menor_var:
          menor_var=fuerza_bruta[i]['parcial']
          nodo_mejor=fuerza_bruta[i]
      print_debug("Solucion:",debug_mode=debug_mode)
      print_debug(fuerza_bruta)
      print_debug(nodo_mejor,debug_mode=debug_mode)
      return nodo_mejor,nodo_cuenta
      
    
    
```

Siendo un arbol de busqueda, la complejidad depende de los niveles y la cantidad de nodos. Los niveles equivalen a la cantidad de grupos que queremos encontrar. Los nodos, dependen de la variable delta. Como hemos establecido, cada nodo tiene **(2*delta+1)** nodos hijo. <br>

Sabemos que el numero de grupos es 9. El codigo de la siquiente celda calcula el numero de nodos para diferente valores de delta. <br>

Como se puede apreciar, la complejidad es de **order exponencial** respecto al numero de grupos. Es decir, el calculo incuye al menos una potencia a la 9. Para valores mayores de 3 ya resulta impracticable.

## Generacion de datos 

La fucion **generar_datos**  en la siguiente celda de codigo genera datos aleatorios segun los argumentos
  - numero_datos: es la cantidad de datos<br>
  - media_datos: es el valor promedio de los datos<br>
  - ds_datos: es la desviacion standard de los datos<br>
  
Para el caso de este problema, trabajamos con **1000 datos**, con una **media de 37** y **desviacion estandard de 2.1**

Los datos son ordenados en fuciones de la variable **peso** y clasificados segun la variable **sexo**.

Podemos ver que el codigo muestra en pantalla una tabla con las caracteristicas de los datos generados.

```
import random
import numpy as np
import pandas as pd
import math

       
def generar_datos(numero_datos=1000,media_datos=37,ds_datos=2.1):
  # Esta funcion general datos aleatorios segun los argumentos
  # numero_datos: es la cantidad de datos
  # media_datos: es el valor promedio de los datos
  # ds_datos: es la desviacion standard de los datos
     
  #Crear dataframe de datos aleatorios, siguiendo la media y desviacion estandard indicadas en el problema
  pesos = [random.gauss(media_datos, math.sqrt(ds_datos*ds_datos)) for _ in range(numero_datos)]
  sexos= [random.choice(['macho','hembra']) for _ in range(numero_datos)]
  ids = [i for i in range(1,numero_datos+1)]
  datos=pd.DataFrame({'Peso':pesos, 'Sexo': sexos, 'Id':ids})
  
  descripcion=[]
  #Mostrar en pantalla muestra de datos 
  print()
  print('Descripcion de datos:')
  descripcion.append({"Cantidad Total":len(datos['Peso']),"Cantidad Hembras":len(datos[datos["Sexo"]=="hembra"]),"Cantidad Machos":len(datos[datos["Sexo"]=="macho"]),"Media":np.mean(datos['Peso']),"Varianza":np.var(datos['Peso']),"Max":np.max(datos['Peso']),"Min": np.min(datos['Peso'])})
  df=pd.DataFrame(descripcion)
  df.index=["Vacas"]
  display(df)
  
  #Ordenar datos por PESO
  datos=datos.sort_values(by=['Peso'])
  
  #Clasificar datos por SEXO
  datos_hembras=datos[datos['Sexo']=='hembra']
  datos_machos=datos[datos['Sexo']=='macho']
    
  return datos_hembras,datos_machos

h,m=generar_datos()
```

## Aplicando el algoritmo

La siguiente celda de codigo contiene la funcion con nombre "agrupar_vacas". Esta funcion hace lo siguiente:

- Ejectua fucion genera_datos para obtener datos a procesar
- Encuentra solucion optima utilizando bucles internos (3 bluces externos, 3 bucles internos) para ambos algoritmos, Fuerza Bruta y Ramificacion y Poda. 
- Encuentra solucion option sin utilizar bucles internos utiliando algoritmo Ramificacion y Poda.
- Presenta los resultados en una tabla
- Presenta los grupos y sus caracteristicas en una tabla

```

def agrupar_vacas(numero_grupos=9, nombre_grupos=["Tratamiento1-Grupo1","Tratamiento1-Grupo2","Tratamiento1-Grupo3","Tratamiento2-Grupo1","Tratamiento2-Grupo2","Tratamiento2-Grupo3","Tratamiento3-Grupo1","Tratamiento3-Grupo2","Tratamiento3-Grupo3"],debug_mode=False,delta=2):
  
  #Generar Datos
  datos_hembras,datos_machos=generar_datos()

  #Resultados
  resultados=[]
  
  #Calcular numero de nodos para caso fuerza bruta
  fb_nodos_totales=0
  for i in range(numero_grupos):
    fb_nodos_totales=fb_nodos_totales+pow(delta*2+1,i)
  fb_nodos_totales=fb_nodos_totales+pow(delta*2+1,numero_grupos-1)

  #Iniciar calculos...
  for vacas,vacas_nombre in zip([datos_hembras,datos_machos],["Vacas Hembra","Vacas Macho"]):
    for poda,modo in zip([True,False],["Ramificacion y Poda","Fuerza Bruta"]):
      for con_bucles_internos,con_bucles_internos_nombre in zip([True,False],["CON Bucles Internos", "SIN Bucles Internos"]):

        #Calcular CON bucles internos
        #Se procede primero a obtener 3 grupos, luego de cada grupo se obtiene 3 grupos mas. Un total de 9 grupos
        if con_bucles_internos:
          #Obtener 3  grupos
          nodo,nodo_cuenta=vaquitas_a_mi(vacas['Peso'].values.tolist(),3,delta=delta,debug_mode=debug_mode,poda=poda,fuerza_bruta=[])     
          nodo['secuencia']=[0]+nodo['secuencia']
          solucion=[0]
          suma_var=0
          #De cada grupo obtenido, obtener 3 grupos mas
          for i in range(len(nodo['secuencia'])-1):
            s,n_c=vaquitas_a_mi(vacas['Peso'][nodo['secuencia'][i]:nodo['secuencia'][i+1]].values.tolist(),3,delta=delta,debug_mode=debug_mode,poda=poda,fuerza_bruta=[])
            #Agregamos los grupos obtenidos para presentar solucion
            nodo_cuenta=nodo_cuenta+n_c
            for i in range(len(s['secuencia'])):
              s['secuencia'][i]=s['secuencia'][i]+solucion[-1]+1
            solucion=solucion+s['secuencia']
            suma_var=suma_var+s['parcial']
        else:
          #Calcular SIN bucles internos
          #Directamente, se procede a obtener 9 grupos.
          if not poda:
            #No calculamos fuerza bruta
            solucion="NA"
            suma_var="NA"
            #nodo_cuenta="Este calculo require de " + str(fb_nodos_totales) + " nodos. IMPOSIBLE"
            nodo_cuenta=fb_nodos_totales
          else:
            #Obtenemos 9 grupos
            nodo,nodo_cuenta=vaquitas_a_mi(vacas['Peso'].values.tolist(),9,delta=delta,debug_mode=debug_mode,poda=True,fuerza_bruta=[])     

        #Almacenamos las soluciones...
        resultados.append({"Tipo de Grupo":vacas_nombre,"Algoritmo":modo,"Bucles Internos":con_bucles_internos_nombre,"Nodos":nodo_cuenta,"Var":suma_var,"Solucion":solucion})
  

  #Mostrar los resultados
  print()
  print('Descripcion de busqueda con diferentes algoritmos:')
  resultados=pd.DataFrame(resultados)
  display(resultados)
  filter_alg=resultados["Algoritmo"]=="Ramificacion y Poda"
  filter_bucle=resultados["Bucles Internos"]=="CON Bucles Internos"
  filter_vacas=resultados["Tipo de Grupo"]=="Vacas Hembra"

  tabla=[]
  for i in range(len(resultados[filter_alg & filter_bucle])):
    sol=resultados[filter_alg & filter_bucle].iloc[i,:]['Solucion']
    grupo_n=resultados[filter_alg & filter_bucle].iloc[i,:]['Tipo de Grupo']
    if grupo_n=="Vacas Hembra":
        vacas=datos_hembras
    else:
        vacas=datos_machos
    for ii,n in zip(range(len(sol)-1),nombre_grupos):
      var=np.var(vacas['Peso'][sol[ii]:sol[ii+1]])
      media=np.mean(vacas['Peso'][sol[ii]:sol[ii+1]])
      cuenta=len(vacas['Peso'][sol[ii]:sol[ii+1]])
      vaca_max=np.max(vacas['Peso'][sol[ii]:sol[ii+1]])
      vaca_min=np.min(vacas['Peso'][sol[ii]:sol[ii+1]])
      tabla.append({"Grupo":n,"Grupo-Categoria":grupo_n,"Grupo-Med":media,"Grupo-Max":vaca_max,"Grupo-Min":vaca_min,"Grupo-Cantidad":cuenta,"Grupo-Var":var})
  tabla=pd.DataFrame(tabla)
  print()
  print('Grupos obtenidos:')
  display(tabla)     
  


agrupar_vacas(delta=10,debug_mode=False)
```
